```python
Migration = Migration.set_index('Country').unstack().reset_index()
```

```python
Migration.merge(Energy_access,on=['level_0','Country'],how='outer').merge(Life_expectancy_at_birth,on=['level_0','Country'],how='outer').merge(GDP_per_capita,on=['level_0','Country'],how='outer').merge(Total_unemployment,on=['level_0','Country'],how='outer')
```

```python
Total = Migration.merge(Energy_access,on=['level_0','Country'],how='outer').merge(Life_expectancy_at_birth,on=['level_0','Country'],how='outer').merge(GDP_per_capita,on=['level_0','Country'],how='outer').merge(Total_unemployment,on=['level_0','Country'],how='outer')
```

```python
Total.columns=["Years","Contry","Migration","Energy_access", "Life_expectancy","GDP_per_capita","Total_unemployment"]
Total
```

```python
import plotly.express as px
```

```python
fig = px.line(Total, x='Years', y='Migration', color='Contry',
              title='Migración por País a lo largo de los Años',
              labels={'Years': 'Año', 'Migration': 'Migración'})

fig.show()
```

```python
fig = px.line(
    Total,
    x='Years',
    y='Total_unemployment',
    color='Contry',
    title='Tasa de Desempleo por Año y País',
    labels={'Years': 'Año', 'Total_unemployment': 'Tasa de Desempleo'},
    line_shape='linear'
)

fig.update_layout(
    xaxis_title='Año',
    yaxis_title='Tasa de Desempleo',
)

fig.show()
```

```python
contry_to_plot = 'Colombia'
filtered_data = Total[Total['Contry'] == contry_to_plot]
fig = px.scatter(
    filtered_data,
    x='GDP_per_capita',
    y='Total_unemployment',
    text='Years',
    title=f'Gráfico de Dispersión: PIB per Capita vs. Tasa de Desempleo Colombia',
    labels={'GDP_per_capita': 'PIB per Capita', 'Total_unemployment': 'Tasa de Desempleo'},
)
fig.show()
```

```python
fig_1 = px.scatter(Total, x="Years", y="GDP_per_capita", color="Contry",
                 size='Total_unemployment', hover_data=['GDP_per_capita'])
fig_1.show()
```
